```python
%pylab inline
import pandas as pd
```

```python
nin_logs = pd.read_csv('nin-cifar10.csv')
maxout_logs = pd.read_csv('maxout-cifar10.csv')
```

```python
nin_logs['error'] = 1 - nin_logs['acc']
maxout_logs['error'] = 1 - maxout_logs['acc']
# We used test set as the validation set. This is okay, since I didn't do any hyperparameter tuning with 
nin_logs['test_error'] = 1 - nin_logs['val_acc']
maxout_logs['test_error'] = 1- maxout_logs['val_acc']
```

```python
nin_logs.plot('epoch', ['error', 'test_error'], figsize=(12, 9), fontsize = 22)
title('Network in network', fontsize = 34)
xlabel('epoch', fontsize = 22)
legend(prop={'size':22})
yticks(arange(0,1.1, 0.1))

# Something is obviously very wrong with my maxout implementation...
maxout_logs.plot('epoch', ['error', 'test_error'], figsize=(12, 9), fontsize = 22)
title('Maxout', fontsize = 34)
xlabel('epoch', fontsize = 22)
legend(prop={'size':22}, loc=4)
yticks(arange(0,1.1, 0.1))
```
